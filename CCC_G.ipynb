```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the entire Excel file with all sheets
file_path = 'World Bank.xlsx'  # Replace with your actual file path

# Load all sheets into DataFrames
Revenue_df = pd.read_excel(file_path, sheet_name='Compliance_Revenue')
Price_df = pd.read_excel(file_path, sheet_name='Compliance_Price')
Emissions_df = pd.read_excel(file_path, sheet_name='Compliance_Emissions')

# Declare the columns to keep for each sheet
columns_to_keep1 = [
    'Name of the initiative',
    'Instrument Type',
    'Jurisdiction Covered',
    2020,
    2021,
    2022,
    2023 
]

columns_to_keep2 = [
    'Name of the initiative',
    'Instrument Type',
    'Jurisdiction Covered',
    'Region',
    'Income group',
    2020,
    2021,
    2022,
    2023 
]

# For Compliance_Emissions, you only need to keep the initiative name and the years 2020–2023
columns_to_keep3 = [
    'Name of the initiative',
    2020, 2021, 2022, 2023  # Keep only these years
]

# Filter the dataframes based on the columns to keep
Revenue_filtered_df = Revenue_df[columns_to_keep1]
Price_filtered_df = Price_df[columns_to_keep2]
Emissions_filtered_df = Emissions_df[columns_to_keep3]

# Exclude the initiative named "EU ETS"
Revenue_filtered_df = Revenue_filtered_df[Revenue_filtered_df['Name of the initiative'] != 'EU ETS']
Price_filtered_df = Price_filtered_df[Price_filtered_df['Name of the initiative'] != 'EU ETS']
Emissions_filtered_df = Emissions_filtered_df[Emissions_filtered_df['Name of the initiative'] != 'EU ETS']

# Merge DataFrames on 'Name of the initiative'
merged_df = Revenue_filtered_df.merge(Price_filtered_df, on='Name of the initiative').merge(Emissions_filtered_df, on='Name of the initiative')

# Select relevant columns for the year 2023 and rename them
data_2023 = merged_df[['Name of the initiative', '2023_x', '2023_y', 2023]]  # '2023_x' for revenue, '2023_y' for price
data_2023.columns = ['Name of the initiative', 'Revenue', 'Price', 'Emissions']  # Renaming columns for clarity

# Clean the data: Convert columns to numeric and coerce errors
data_2023['Revenue'] = pd.to_numeric(data_2023['Revenue'], errors='coerce')
data_2023['Price'] = pd.to_numeric(data_2023['Price'], errors='coerce')
data_2023['Emissions'] = pd.to_numeric(data_2023['Emissions'], errors='coerce')

# Drop rows with NaN values
data_2023.dropna(subset=['Revenue', 'Price', 'Emissions'], inplace=True)


# Create scatter plots to visualize the correlations for top half initiatives
plt.figure(figsize=(14, 6))

# Correlation between Emissions and Price
plt.subplot(1, 2, 1)
sns.regplot(data=data_2023, x='Price', y='Emissions', scatter_kws={'color': 'blue', 'alpha': 0.6}, line_kws={'color': 'red'})
plt.title('Correlation Between Emissions and Price (2023)', fontsize=14)
plt.xlabel('Price', fontsize=12)
plt.ylabel('Emissions', fontsize=12)

# Correlation between Emissions and Revenue
plt.subplot(1, 2, 2)
sns.regplot(data=data_2023, x='Revenue', y='Emissions', scatter_kws={'color': 'blue', 'alpha': 0.6}, line_kws={'color': 'red'})
plt.title('Correlation Between Emissions and Revenue (2023)', fontsize=14)
plt.xlabel('Revenue', fontsize=12)
plt.ylabel('Emissions', fontsize=12)

plt.tight_layout()
plt.show()
```

## Convert Format of Numbers

```python
import pandas as pd

# Load the Excel file
file_path = "Emission by Industry.xlsx"
sheets = ["2023", "2022", "2021"]

# Process each sheet
for sheet in sheets:
    # Read the sheet into a DataFrame
    df = pd.read_excel(file_path, sheet_name=sheet)

    # Loop through each relevant column for emissions and profit
    for col in df.columns:
        # Clean numeric columns for emissions and scope
        if 'emission' in col.lower() or 'scope' in col.lower():
            df[col] = df[col].astype(str).str.replace(r'[\$,]', '', regex=True)  # Remove $ and ,
            df[col] = df[col].str.replace(r'\s+', '', regex=True)  # Remove spaces
            df[col] = pd.to_numeric(df[col], errors='coerce')  # Convert to numeric

    # Clean specific profit columns
    if sheet == "2023":
        profit_col = "2022 Profit (Millions USD)"
        revenue_col = "2022 Sales / Revenue (Millions USD)"
    elif sheet == "2022":
        profit_col = "2021 Profit (Million USD)"
        revenue_col = "2021 Sales / Revenue (Million USD)"
    elif sheet == "2021":
        profit_col = "Profit"
        revenue_col = "Sales/Revenue"
    
    # Clean the specific profit and revenue columns
    for col in [profit_col, revenue_col]:
        if col in df.columns:
            df[col] = df[col].astype(str).str.replace(r'[\$,]', '', regex=True)  # Remove $ and ,
            df[col] = df[col].str.replace(r'\s+', '', regex=True)  # Remove spaces
            df[col] = pd.to_numeric(df[col], errors='coerce')  # Convert to numeric

    # Save the cleaned DataFrame back to a new Excel file
    output_file = f"Cleaned_Emission_by_Industry_{sheet}.xlsx"
    df.to_excel(output_file, index=False)

print("Data cleaned and saved for each sheet.")
```
