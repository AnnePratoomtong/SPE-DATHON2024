```python
import pandas as pd


# Example: Select 'Country' and '2023' column from each dataset
GHGEnergy_2023 = GHGEnergy_df[['Region/Country/Economy', 2023]].rename(columns={'Region/Country/Economy': 'Country', 2023: 'Energy Emissions (2023)'})
GHGFugitive_2023 = GHGFugitive_df[['Region/Country/Economy', 2023]].rename(columns={'Region/Country/Economy': 'Country', 2023: 'Fugitive Emissions (2023)'})
GHGFuel_2023 = GHGFuel_df[['Region/Country/Economy', 2023]].rename(columns={'Region/Country/Economy': 'Country', 2023: 'Fuel Emissions (2023)'})
GHGMarine_2023 = GHGMarine_df[['Region/Country/Economy', 2023]].rename(columns={'Region/Country/Economy': 'Country', 2023: 'Marine Emissions (2023)'})
GHGAviation_2023 = GHGAviation_df[['Region/Country/Economy', 2023]].rename(columns={'Region/Country/Economy': 'Country', 2023: 'Aviation Emissions (2023)'})
GHGperTES_2023 = GHGperTES_df[['Region/Country/Economy', 2023]].rename(columns={'Region/Country/Economy': 'Country', 2023: 'CO2 per TES (2023)'})
GHGperPop_2023 = GHGperPop_df[['Region/Country/Economy', 2023]].rename(columns={'Region/Country/Economy': 'Country', 2023: 'CO2 per Population (2023)'})

# Step 2: Merge them all together on the 'Country' column
merged_2023_df = GHGEnergy_2023.merge(GHGFugitive_2023, on='Country', how='inner')\
                               .merge(GHGFuel_2023, on='Country', how='inner')\
                               .merge(GHGMarine_2023, on='Country', how='inner')\
                               .merge(GHGAviation_2023, on='Country', how='inner')\
                               .merge(GHGperTES_2023, on='Country', how='inner')\
                               .merge(GHGperPop_2023, on='Country', how='inner')

# Display the final merged dataframe for 2023 CO2 emissions by different sources
display(merged_2023_df)

# Save the result to a CSV file if needed
merged_2023_df.to_csv('2023_co2_emissions_by_source.csv', index=False)
```

```python
# read excel file GHGemission.xlxs into dataframe name emission_df

emissions_df = pd.read_excel('/content/drive/MyDrive/SPEDATATHON/GHGemission.xlsx')
```

```python
import matplotlib.pyplot as plt

# Assuming emissions_df has years as columns starting from the second column
# and the first column is just a descriptor ('million tonnes of CO2 eq')
# Also assuming the OWID temperature data is loaded as owid_df

# Extract years (columns starting from the second one)
years = emissions_df.columns[1:].astype(int)

# Extract emissions data from rows corresponding to 'Energy Emissions', 'Fugitive Emissions', and 'Fuel Emissions'
# and columns starting from the second one
energy_emissions = emissions_df.iloc[0, 1:]  # First row is 'Energy Emissions'
fugitive_emissions = emissions_df.iloc[1, 1:]  # Second row is 'Fugitive Emissions'
fuel_emissions = emissions_df.iloc[2, 1:]  # Third row is 'Fuel Emissions'

# Extract temperature change data from the OWID dataset for the same years (1971-2022)
ghg_temp_change = owid_df[owid_df['year'].isin(years)].groupby('year')['temperature_change_from_ghg'].mean()

# Create the figure and axis objects
fig, ax1 = plt.subplots(figsize=(10, 6))

# Plot CO2 emissions (Energy, Fugitive, Fuel) on the left y-axis
ax1.set_xlabel('Year')
ax1.set_ylabel('CO2 Emissions (Million tonnes)', color='blue')

# Plot energy emissions
ax1.plot(years, energy_emissions.values, label='Energy Emissions', color='green', linewidth=2)
# Plot fugitive emissions
ax1.plot(years, fugitive_emissions.values, label='Fugitive Emissions', color='orange', linewidth=2)
# Plot fuel emissions
ax1.plot(years, fuel_emissions.values, label='Fuel Emissions', color='red', linewidth=2)

# Add a legend for the left y-axis
ax1.legend(loc='upper left')

# Create the second y-axis (right side) for temperature change
ax2 = ax1.twinx()
ax2.set_ylabel('Temperature Change (°C)', color='purple')

# Plot temperature change from GHG (right y-axis)
ax2.plot(ghg_temp_change.index, ghg_temp_change.values, label='Temperature Change (GHG)', color='purple', linewidth=2)

# Add a legend for the right y-axis
ax2.legend(loc='upper right')

# Add a title and grid
plt.title('Global CO2 Emissions and Temperature Change Over Time (1971-2022)')
plt.grid(True)

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Create subplots: one for each emission source and two additional for CO2 per TES and CO2 per POP
fig, axes = plt.subplots(5, 1, figsize=(10, 15), sharex=True)

# Plot energy emissions on its own scale
axes[0].set_ylabel('Energy Emissions (Million tonnes)', color='green')
axes[0].plot(years, emissions_df.iloc[0, 1:], label='Energy Emissions', color='green', linewidth=2)
axes[0].legend(loc='upper left')
axes[0].grid(True)

# Plot fugitive emissions on its own scale
axes[1].set_ylabel('Fugitive Emissions (Million tonnes)', color='orange')
axes[1].plot(years, emissions_df.iloc[1, 1:], label='Fugitive Emissions', color='orange', linewidth=2)
axes[1].legend(loc='upper left')
axes[1].grid(True)

# Plot fuel emissions on its own scale
axes[2].set_ylabel('Fuel Emissions (Million tonnes)', color='red')
axes[2].plot(years, emissions_df.iloc[2, 1:], label='Fuel Emissions', color='red', linewidth=2)
axes[2].legend(loc='upper left')
axes[2].grid(True)

# Plot CO2 per TES on its own scale (row 7 in emissions_df)
axes[3].set_ylabel('CO2 per TES', color='blue')
axes[3].plot(years, emissions_df.iloc[6, 1:], label='CO2/TES', color='blue', linewidth=2)
axes[3].legend(loc='upper left')
axes[3].grid(True)

# Plot CO2 per POP on its own scale (row 9 in emissions_df)
axes[4].set_xlabel('Year')
axes[4].set_ylabel('CO2 per POP', color='purple')
axes[4].plot(years, emissions_df.iloc[8, 1:], label='CO2/POP', color='purple', linewidth=2)
axes[4].legend(loc='upper left')
axes[4].grid(True)

# Add temperature change to all subplots (right y-axis)
for ax in axes:
    ax_temp = ax.twinx()
    ax_temp.set_ylabel('Temperature Change (°C)', color='gray')
    ax_temp.plot(ghg_temp_change.index, ghg_temp_change.values, label='Temperature Change (GHG)', color='gray', linewidth=2)
    ax_temp.legend(loc='upper right')

plt.suptitle('Global CO2 Emissions, CO2 per TES, CO2 per POP, and Temperature Change Over Time (1971-2022)', y=0.95)
plt.show()
```
